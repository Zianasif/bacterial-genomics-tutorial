# Bacterial Genomics Tutorial Walkthrough

This notebook shows a minimal, end-to-end walkthrough using the run profiles.

Prerequisites:
- A bash shell (Linux, macOS, or WSL)
- Conda environment created from environment.yaml
- Pipeline scripts made executable (chmod +x *.sh)

```python
import os
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / 'run_pipeline.sh').exists() and (repo_root.parent / 'run_pipeline.sh').exists():
    repo_root = repo_root.parent

os.environ['REPO_ROOT'] = str(repo_root)
print(f'Repo root: {repo_root}')
```

```python
%%bash
set -euo pipefail
cd "$REPO_ROOT"
./run_pipeline.sh --profile demo
```

```python
%%bash
set -euo pipefail
cd "$REPO_ROOT"
./run_pipeline.sh --profile full --skip-download
```

```python
from pathlib import Path

runs_dir = Path(os.environ.get('REPO_ROOT', '.')) / 'results'
if runs_dir.exists():
    run_dirs = sorted([p.name for p in runs_dir.iterdir() if p.is_dir()])
    print('Runs:')
    for name in run_dirs:
        print(' -', name)
else:
    print('No results directory yet.')
```

## Output layout

Each run creates a run folder under the results directory. Raw outputs stay in the work subfolder. The organized outputs tree collects: 
- QC reports
- Trimmed reads
- Assembly outputs and polishing artifacts
- Reordered contigs and typing results
- Annotation outputs and stats
- Pangenome analysis outputs and summary plots
